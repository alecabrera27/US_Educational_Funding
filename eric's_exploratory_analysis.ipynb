```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.linear_model import LinearRegression
```

```python
df2 = df1[['STATE',
           'TOTAL_REVENUE',
           'ENROLL',
           'INSTRUCTION_EXPENDITURE',
           'SUPPORT_SERVICES_EXPENDITURE',
           'GRADES_ALL_G',
           'AVG_MATH_4_SCORE',
           'AVG_READING_4_SCORE'
          ]]
df2.head()
```

## Plotting Data for Average 4th Grade Math Scores

```python
#Looking for relationship between total state revenue and average 4th Grade Math Scores
plt.scatter(df2['TOTAL_REVENUE'], df2['AVG_MATH_4_SCORE'])
plt.xlabel('Total State Revenue')
plt.ylabel('AVG_MATH_4_SCORE')
plt.show()
```

```python
#Total Rev per student
plt.scatter(df2['TOTAL_REVENUE']/df2['ENROLL'], df2['AVG_MATH_4_SCORE'])
plt.xlabel('Total State Revenue per Student')
plt.ylabel('AVG_MATH_4_SCORE')
plt.show()
```

```python
#Looking for relationship between enrollment and average 4th grade math scores
plt.scatter(df2['ENROLL'], df2['AVG_MATH_4_SCORE'])
plt.xlabel('Total State Enrollment')
plt.ylabel('AVG_MATH_4_SCORE')
plt.show()
```

```python
#Looking for relationship between Instructor spending and average 4th grade math scores
plt.scatter(df2['INSTRUCTION_EXPENDITURE'], df2['AVG_MATH_4_SCORE'])
plt.xlabel('INSTRUCTION Spending')
plt.ylabel('AVG_MATH_4_SCORE')
plt.show()
```

```python
# instructor spending per student
plt.scatter(df2['INSTRUCTION_EXPENDITURE']/df2['ENROLL'], df2['AVG_MATH_4_SCORE'])
plt.xlabel('INSTRUCTION Spending per Student')
plt.ylabel('AVG_MATH_4_SCORE')
plt.show()
```

```python
#Looking for relationship between support services average 4th grade math scores
plt.scatter(df2['SUPPORT_SERVICES_EXPENDITURE'], df2['AVG_MATH_4_SCORE'])
plt.xlabel('Support Services Spending')
plt.ylabel('AVG_MATH_4_SCORE')
plt.show()
```

```python
#support services per student
plt.scatter(df2['SUPPORT_SERVICES_EXPENDITURE']/df2['ENROLL'], df2['AVG_MATH_4_SCORE'])
plt.xlabel('Support Services Spending per Student')
plt.ylabel('AVG_MATH_4_SCORE')
plt.show()
```

```python
#Looking for relationship between Instructor expenditure per student and average 4th grade math scores
plt.scatter(df2['INSTRUCTION_EXPENDITURE']/df2['ENROLL'], df2['AVG_MATH_4_SCORE'])
plt.xlabel('Instruction Spending per Student')
plt.ylabel('AVG_MATH_4_SCORE')
plt.show()
```
